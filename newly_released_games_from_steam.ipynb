```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url = 'https://store.steampowered.com/search/?sort_by=Released_DESC&supportedlang=english'
page = requests.get(url)
soup = BeautifulSoup(page.text,'html')
```

```python
header1 = ["Title", "Released Date", "Discount", "Final Price"]
new_games_table = pd.DataFrame(columns=header1)
```

```python
sales_table = soup.find('div', id='search_resultsRows')
entries = sales_table.find_all('a')
new_games_table = new_games_table[0:0]
no_of_entries = 0
for entry in entries:
    line = []
    line.append(entry.find('span').text.strip())
    line.append(entry.find('div', class_="col search_released responsive_secondrow").text.strip())
    # review = entry.find('div', class_="col search_reviewscore responsive_secondrow").text.strip()
    # if review:
    #     line.append(entry.find('div', class_="col search_reviewscore responsive_secondrow").text.strip())
    # else:
    #     line.append('')
    isDiscounted = entry.find('div', class_="discount_pct")
    if isDiscounted:
        line.append(entry.find('div', class_="discount_pct").text.strip())
    else:
        line.append('')
    line.append(entry.find('div', class_='discount_final_price').text.strip())
    
    try:
        curr_len = len(new_games_table)
        new_games_table.loc[curr_len]=line
    except ValueError:
        print("Mismatch Column with:", line)

new_games_table
```
