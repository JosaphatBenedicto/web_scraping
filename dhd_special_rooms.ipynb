```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url = 'https://kairosoft.fandom.com/wiki/Special_Rooms_(Dream_House_Days)'
page = requests.get(url)
soup = BeautifulSoup(page.text,'html')
table = soup.find('table', class_ = "wikitable oddrow sortable")
```

```python
table_header = table.find_all('th')
table_header = [header.text.strip() for header in table_header]
dhd_special_room = pd.DataFrame(columns=table_header)
```

```python
table_data = table.find_all('tr')
for data_row in table_data[1:]:
    data_row = data_row.find_all('td')
    entry = [data.text.strip() if data.text.strip()!='∅' else '' for data in data_row ]
    
    curr_len = len(dhd_special_room)
    dhd_special_room.loc[curr_len] = entry
    
dhd_special_room
```
